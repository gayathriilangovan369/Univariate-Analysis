```python
import pandas as pd
```

```python
dataset=pd.read_csv("kidney_disease.csv")
```

```python
~checking the logic to automate the columns based on the datatypes
quan=[]
qual=[]
for ColumnName in dataset.columns:
    if (dataset[ColumnName]).dtypes=='O':
        qual.append(ColumnName)
    else:
        quan.append(ColumnName)
```

```python
#Implementing that as functions
def quanqual(dataset):
    quan=[]
    qual=[]
    for ColumnName in dataset.columns:
        if (dataset[ColumnName]).dtypes=='O':
            qual.append(ColumnName)
        else:
            quan.append(ColumnName)
    return quan,qual
```
